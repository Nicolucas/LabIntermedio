# Sr90 - Datos Sin Magnetos

#### Carga de datos...

```python
data = loadtxt("TI204M1")
phi0 = data[:,0]
R0 = data[:,1]
R1 = data[:,2]
num=shape(phi0)[0]
dataMatrix=matrix([phi0,R0,R1])
m=dataMatrix.T

col = 0
datLab=m[np.array(m[:,col].argsort(axis=0).tolist()).ravel()]
```

#### organizar datos

```python
t = linspace(-pi/2, pi/2, num)
r0=array(datLab[:,1])
r1=array(datLab[:,2])
```

```python
# polar plot using add_axes and polar projection
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
ax.plot(t, r0);
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, .6,.6], polar=True)
ax.plot(t, r1);
```
